```python
import matplotlib.pyplot as plt
import json
```

```python
with open('myKV_cuckoo.json', 'r') as f:
    d = json.load(f)
```

```python
tableNumList1S = ['myKV_cuckoo', 'myKV_simple', 'myKV_hopscotch', 'ycsbc']
tableNumList2S = ['myKV2S_cuckoo', 'myKV2S_simple', 'myKV2S_hopscotch', 'ycsbc']
tableNumList1Sn = ['myKV_cuckoo', 'myKV_simple', 'myKV_hopscotch']
tableNumList2Sn = ['myKV2S_cuckoo', 'myKV2S_simple', 'myKV2S_hopscotch']
```

```python
def drawLat(fnameList, wl):
    plt.figure(figsize=(10,6))
    for fname in fnameList:
        _fname = fname + '_lat.json'
        f = open(_fname, 'r')
        d = json.load(f)
        f.close()
        dx = d[wl]['tnum']
        dy = [i / 1000 for i in  d[wl]['tput']]
        plt.plot(dx[:10], dy[:10])
    plt.legend(['cuckoo', 'simple', 'hopscotch', 'redis'], fontsize=28, loc=4)
    plt.xticks(fontsize=24)
    plt.yticks(fontsize=24)
    plt.xlabel("Num Threads", fontsize=28)
    plt.ylabel("Latency (us)", fontsize=28)
    plt.savefig('2S_{}_lat.pdf'.format(wl), bbox='tight')
```

```python
def draw(fnameList, wl):
    plt.figure(figsize=(10,6))
    for fname in fnameList:
        _fname = fname + '.json'
        f = open(_fname, 'r')
        d = json.load(f)
        f.close()
        dx = d[wl]['tnum']
        dy = d[wl]['tput']
        plt.plot(dx[:10], dy[:10])
    plt.legend(['cuckoo', 'simple', 'hopscotch', 'redis'], fontsize=28, loc=4)
    plt.xticks(fontsize=24)
    plt.yticks(fontsize=24)
    plt.xlabel("Num Threads", fontsize=28)
    plt.ylabel("Throughput (Kops)", fontsize=28)
    plt.savefig('1S_{}.pdf'.format(wl), bbox='tight')
```

```python
drawLat(tableNumList1Sn, 'D')
```

```python
draw(tableNumList1S, 'E')
```
