```python
import matplotlib.pyplot as plt
import json
```

```python
with open('myKV_cuckoo.json', 'r') as f:
    d = json.load(f)
```

```python
tableNumList1S = ['myKV_cuckoo', 'myKV_simple', 'myKV_hopscotch', 'ycsbc']
tableNumList2S = ['myKV2S_cuckoo', 'myKV2S_simple', 'myKV2S_hopscotch', 'ycsbc']
```

```python
def draw(fnameList, wl):
    for fname in fnameList:
        _fname = fname + '.json'
        f = open(_fname, 'r')
        d = json.load(f)
        f.close()
        dx = d[wl]['tnum']
        dy = d[wl]['tput']
        plt.plot(dx, dy)
    plt.legend(['cuckoo', 'simple', 'hopscotch', 'redis'])
    plt.xlabel("Num Threads")
    plt.ylabel("Throughput (Kops)")
    plt.savefig('1S_{}.pdf'.format(wl), bbox='tight')
```

```python
draw(tableNumList1S, 'E')
```

```python
draw(tableNumList2S, 'A')
```
